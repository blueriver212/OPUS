```python
import csv
import pandas as pd
```

```python
data = pd.read_csv('UCS_Satellite_db.csv')

data.head()
```

```python
# print each column name, the data type and an example
for col in data.columns:
    print(f"Column: {col}")
    print(f"Data Type: {data[col].dtype}")
    print(f"Example: {data[col].iloc[0]}")
    print()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = data.copy()

# Ensure numeric mass column is usable before filtering
df['Launch Mass (kg.)'] = pd.to_numeric(df['Launch Mass (kg.)'], errors='coerce')
initial_count = len(df)

# --- 1. Filter for LEO class of orbit ---
leo_filter = df['Class of Orbit'].str.strip().str.upper() == 'LEO'
df = df[leo_filter]
removed = initial_count - len(df)
print(f"Removed {removed} satellites not in LEO")

# --- 2. Remove SpaceX and OneWeb ---
before = len(df)
df = df[~df['Operator/Owner'].isin(['SpaceX', 'Spacex', 'OneWeb Satellites'])]
removed = before - len(df)
print(f"Removed {removed} SpaceX or OneWeb satellites")

# --- 3. Filter out government/civil/military users ---
before = len(df)
df = df[~df['Users'].str.contains('Government|Military|Civil', case=False, na=False)]
removed = before - len(df)
print(f"Removed {removed} government or civil satellites")

# --- 4. Remove CubeSat-sized satellites (mass ≤ 10 kg) ---
before = len(df)
df = df[df['Launch Mass (kg.)'] > 10]
removed = before - len(df)
print(f"Removed {removed} CubeSat-sized satellites (mass ≤ 10 kg)")

# --- 5. Keep only Commercial users ---
before = len(df)
df = df[df['Users'].str.contains('Commercial', case=False, na=False)]
removed = before - len(df)
print(f"Removed {removed} non-commercial satellites")

# --- Final result ---
print(f"Final satellite count: {len(df)}")

# Print the mean mass and radius of the filtered dataset
mean_mass = df['Launch Mass (kg.)'].mean()
print(f"Mean Launch Mass (kg): {mean_mass:.2f}")

# --- Bar chart of mass distribution ---
mass_counts = df['Launch Mass (kg.)'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
mass_counts.plot(kind='bar')
plt.xlabel('Launch Mass (kg)')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Launch Mass (Filtered Dataset)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Bar chart of Class of Orbit ---
type_of_orbit = df['Type of Orbit'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
type_of_orbit.plot(kind='bar')
plt.xlabel('Type of Orbit')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Types of Orbit (Filtered Dataset)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Bar chart of Operator/Owner (over 10 sats) ---
ownership_counts = df['Operator/Owner'].value_counts()
ownership_counts = ownership_counts[ownership_counts > 10].sort_values(ascending=False)
plt.figure(figsize=(12, 6))
ownership_counts.plot(kind='bar')
plt.xlabel('Operator/Owner')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Operators/Owners (Filtered Dataset, >10 Satellites)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Bar chart of Purpose ---
purpose_counts = df['Purpose'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
purpose_counts.plot(kind='bar')
plt.xlabel('Purpose')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Purposes (Filtered Dataset)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Start with a fresh copy and ensure mass is numeric
df_cubesat = data.copy()
df_cubesat['Launch Mass (kg.)'] = pd.to_numeric(df_cubesat['Launch Mass (kg.)'], errors='coerce')

# --- Filter: CubeSats only (mass ≤ 10 kg) ---
df_cubesat = df_cubesat[df_cubesat['Launch Mass (kg.)'] <= 10]
print(f"Number of CubeSats (≤ 10 kg): {len(df_cubesat)}")

# --- Bar chart: Top CubeSat Operators ---
operator_counts = df_cubesat['Operator/Owner'].value_counts().sort_values(ascending=False)
print(operator_counts)
plt.figure(figsize=(12, 6))
operator_counts[operator_counts > 5].plot(kind='bar')  # Only show if > 5 CubeSats for clarity
plt.xlabel('Operator/Owner')
plt.ylabel('Number of CubeSats')
plt.title('Top CubeSat Operators/Owners (>5 Satellites)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Bar chart of Class of Orbit ---
type_of_orbit = df['Type of Orbit'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
type_of_orbit.plot(kind='bar')
plt.xlabel('Type of Orbit')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Types of Orbit (Filtered Dataset)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Bar chart: Orbital Class Distribution of CubeSats ---
orbit_class_counts = df_cubesat['Class of Orbit'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
orbit_class_counts.plot(kind='bar')
plt.xlabel('Class of Orbit')
plt.ylabel('Number of CubeSats')
plt.title('CubeSat Orbital Class Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
